```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("bio-degradabale-data.csv", sep=';', header=None)
data.shape
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
#Assinging features and dependent variables
X = data.iloc[:, 0:41]
y = data[41]
```

```python
X.head()
```

```python
y.unique()
```

```python
#Scaling
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
X_scaled = pd.DataFrame(scalar.fit_transform(X))
X_scaled.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=40, test_size=.30)
```

```python
#Fitting the Logistic Regression
logReg = LogisticRegression()
logReg.fit(X_train, y_train)
y_pred = logReg.predict(X_test)

print(metrics.accuracy_score(y_pred, y_test))
```

```python
#Applying CrossValidation with CV =10
print(cross_val_score(LogisticRegression(), X_scaled, y, cv=10, scoring='accuracy').mean())
```
