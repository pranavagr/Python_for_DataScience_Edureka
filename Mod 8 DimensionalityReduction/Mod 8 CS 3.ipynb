```python
"""1. We shall use the same dataset used in previous assignment - digits. Make a 80-20 train/test split.
[Hint: Explore datasets module from scikit learn]""" 

#Loading the dataset
data = pd.read_csv("breast-cancer-data.csv", header=0)
data.shape
```

```python
#Dropping the diagnosis column to calculate the PC's
data1 = data.drop(["diagnosis"], axis=1)
```

```python
data1.head()
```

```python
#Reducing the principal components to 2

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(data1)
transformed_data = pca.transform(data1)
```

```python
#Creating new dataframe

new_df = pd.DataFrame(transformed_data)
new_df.head()
```

```python
#Setting index, column names and the target vector
new_df.index = data1.index     
new_df.columns = ["PC1", "PC2"] #Setting column names
new_df["diagnosis"] = data["diagnosis"]
new_df.head()
```
